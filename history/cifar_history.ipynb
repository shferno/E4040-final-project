# Fishnet & Cifar 

Fishnet 55 & Fishnet 77 using Dataset Cifar_10 & Cifar_100

## Fishnet 55 & Cifar_10

```python
import pickle
import os
from matplotlib import pyplot as plt
```

```python
path = './fishnet55-cifar10-15_58_08'
with open(path, 'rb') as f:
    h1 = pickle.load(f)
```

```python
h1
```

```python
max(h1['val_accuracy'])
```

```python
plt.figure(dpi=300)
plt.plot(h1['accuracy'])
plt.plot(h1['val_accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.ylim([0, 1])
plt.legend(['train_accuracy', 'validation_accuracy'], 
           loc='upper left')
plt.show()
```
